## 1. Dataset Selection


### Descriptions for Each Column in the Dataset:

- **Job Id:** A unique identifier for each job posting.
- **Experience:** The required or preferred years of experience for the job.
- **Qualifications:** The educational qualifications needed for the job.
- **Salary Range:** The range of salaries or compensation offered for the position.
- **Location:** The city or area where the job is located.
- **Country:** The country where the job is located.
- **Latitude:** The latitude coordinate of the job location.
- **Longitude:** The longitude coordinate of the job location.
- **Work Type:** The type of employment (e.g., full-time, part-time, contract).
- **Company Size:** The approximate size or scale of the hiring company.
- **Job Posting Date:** The date when the job posting was made public.
- **Preference:** Special preferences or requirements for applicants (e.g., Only Male or Only Female, or Both).
- **Contact Person:** The name of the contact person or recruiter for the job.
- **Contact:** Contact information for job inquiries.
- **Job Title:** The job title or position being advertised.
- **Role:** The role or category of the job (e.g., software developer, marketing manager).
- **Job Portal:** The platform or website where the job was posted.
- **Job Description:** A detailed description of the job responsibilities and requirements.
- **Benefits:** Information about benefits offered with the job (e.g., health insurance, retirement plans).
- **Skills:** The skills or qualifications required for the job.
- **Responsibilities:** Specific responsibilities and duties associated with the job.
- **Company Name:** The name of the hiring company.
- **Company Profile:** A brief overview of the company's background and mission.

## Sampling method

We use the sampling method to avoid crashing in the colab while they are runnning. Sampling will take only a small percentage of the dataset (which in our case is 50%) to be used in the calculation and processing the dataset,

```python
import random

n = len(df)  # Calculate the number of rows in the DataFrame
s = n // 2  # Sample size of 50%
skip = random.sample(range(n), n - s)
sample = df.drop(skip)
```

###Data Visualisation


```python
plt.figure(figsize=(12,6))
sns.countplot(x='Qualifications', data= sample, palette='viridis')
plt.title('Distribution of Job by Qualifications')
plt.xlabel('Qualifications')
plt.ylabel('Number of Job')
plt.xticks(rotation=45)
plt.show()
```

In summary, the code generates a count plot to visualize the distribution of jobs based on different qualifications, providing a quick overview of the frequency of each qualification category in the dataset. The color palette 'viridis' is chosen for aesthetic appeal, and the plot is customized with titles, axis labels, and rotation for better presentation.

```python
plt.figure(figsize=(14, 6))
sns.histplot(sample['Company Size'], bins=20, kde=True, color='blue')
plt.title('Distribution of Company Size')
plt.xlabel('Number of Company Size')
plt.ylabel('Frequency')
plt.show()
```

The code generates a histogram to visualize the distribution of company sizes in the dataset. The histogram provides insights into the frequency or count of different company size ranges, and the addition of a kernel density estimate (KDE) provides a smoothed representation of the data distribution. The customization with a title and axis labels enhances the interpretability of the plot.

```python
plt.figure(figsize=(10, 8))
sns.heatmap(sample.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

The code generates a heatmap to visually represent the correlation matrix of numerical columns in the sample DataFrame. The color intensity indicates the strength and direction of the correlation between pairs of variables, while the numerical annotations provide precise correlation values. This type of visualization is helpful for understanding relationships between variables and identifying patterns in the dataset.

```python
sns.boxplot(x='Work Type', y='Company Size', data=sample, palette='pastel')
```

The boxplot visualizes the distribution of 'Company Size' for different categories of 'Work Type'. The boxes show the interquartile range (IQR) of the data, the horizontal line inside each box represents the median, and the whiskers extend to the minimum and maximum values within 1.5 times the IQR. This type of plot is useful for comparing the distribution of a numerical variable across different categories.
