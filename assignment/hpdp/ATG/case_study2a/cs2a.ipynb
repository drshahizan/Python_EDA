# Implementation of Automated EDA Tools

8. Firstly, we need to import the sweetviz module  to get the access of the functionalities.

```python
import sweetviz as sv
```

####In creating the report, we able to use SweetViz's three main function:
    (1) analyze()
    (2) compare()
    (3) compare_intra()

In this case study, we will only be using two functions which are analyze() and compare_intra() since we have one dataset only.



 ### 9. Analyze()
    
    Use analyze() function combining with the show_html() function in order to analyze our dataframe

```python
atg_report = sv.analyze(df)
atg_report.show_notebook()
```

Here, we want to download our report and review the result in new browser tab.

```python
from google.colab import files

atg_report.show_html()
files.download('SWEETVIZ_REPORT.html')
```

####There are optional arguments in the analyze() function:
    (1) target_feat
    (2) feat_cfg
    (3) pairwise_analysis

In this case study, we only use two optional arguments which are *target_feat* and *pairwise_analysis*.

 10. Target Feature Analysis (*target_feat argument*)
  
    Please note that boolean and numerical datatype column only can be used in this analysis. Based on this case study, I will use the *population* column to be as the target.

```python
report_target_analysis = sv.analyze(df, target_feat='population')

#To display it in other new browser tab
report_target_analysis.show_html('Target_Analysis.html')
files.download('Target_Analysis.html')
```

11. Pairwise Analysis (*pairwise_analysis argument*)

    This analysis is to examining relationships between pairs of features in the dataset through the visualisation and statistics which will help in exploring and intepreting these relationships.

For this analysis, I set the setting "*auto*" to help manage by automatically in deciding whether to perform pairwise analysis based on the number of features.



```python
report_pairwise_analysis = sv.analyze(df, pairwise_analysis='auto')
report_pairwise_analysis.show_html('Pairwise_Analysis.html')

files.download('Pairwise_Analysis.html')
```

 ### 12. Compare_intra()
    
    Use compare_intra() function to compare two subsets( in our vaccination district dataset.

```python
!pip install --upgrade sweetviz
```

```python
atg_report = sv.compare_intra(df, df["sex"] == "female", ["male", "female"], "population")

atg_report.show_html()
files.download('SWEETVIZ_REPORT.html')
```

### 13. Show the report
      Here, we can choose to use show_html() function to display our report analysis in other new browser tab. Other method, show_notebook() can be used also to display in current tab(Google Colab).

```python
atg_report.show_html( filepath='SWEETVIZ_REPORT.html',
            open_browser=True,
            layout='widescreen',
            scale=None)

files.download('SWEETVIZ_REPORT.html')
```

```python
atg_report.show_notebook(  w=None,
                h=None,
                scale=None,
                layout='widescreen',
                filepath=None)
```
