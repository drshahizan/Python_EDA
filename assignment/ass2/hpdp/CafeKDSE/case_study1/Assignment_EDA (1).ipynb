## Questions and Answers (Q&A)





#### Q1: What is the overall distribution of comorbidity cases(partial, full, and booster) in the dataset?  


As we can see from pie chart and calculations, partial_comorbid has the highest values of overall distribution of comorbidity cases.

```python
#Calculate sum of comorbitidy cases for each type
comorbidity_distribution= data[['partial_comorbid','full_comorbid','booster_comorbid']].sum()
print("Total sum of each comorbidity cases:")
comorbidity_distribution
```

```python
#Plot a bar chart to visualize the distribution
comorbidity_distribution.plot(kind='pie', autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p * sum(comorbidity_distribution)/100), colors=['blue', 'orange', 'red'])
plt.title('Distribution of Comorbidity Cases')
plt.ylabel('')
plt.show()
```

#### Q2: How does the distribution of comorbidity cases vary across different states?

From the stacked bar chart, we can observe that Kuala Lumpur has the highest comorbidity cases, meanwhile W.P. Labuan has the least value of cases

```python
#Group by 'state' and sum of comorbidity cases for each type
comorbidity_by_state= data.groupby("state")[['partial_comorbid', 'full_comorbid', 'booster_comorbid']].sum()
comorbidity_by_state
```

```python
#plot stacked bar chart to compare the distribution across states
comorbidity_by_state.plot(kind='bar', stacked=True)
plt.title('Comorbidity Case Distribution Across States ')
plt.xlabel('State')
plt.ylabel('Number of Cases')
plt.show()
```

#### Q3: What is average number of comorbidity cases per district

W.P Kuala Lumpur has the highest values of average number of comorbidity cases. This result is influenced on the result above, as it has the highest value of comorbidity cases, therefore the average value also will be high

```python
#Create new column for total comorbidity case in each district
data['total_comorbid'] = data[['partial_comorbid','full_comorbid','booster_comorbid']].sum(axis=1)

#Calculate average number of comorbidity cases per district
average_comorbid_per_state = data.groupby('state')['total_comorbid'].mean()
average_comorbid_per_state
```

```python
#plot bar chart
average_comorbid_per_state.plot(kind='bar', color='red')
plt.title('Average Comorbidiy Cases per State')
plt.xlabel('State')
plt.ylabel('Average Number of Cases')
plt.show
```

#### Q4: Is there any correlation between number of healthcare worker(HCW) cases and comorbidity cases

Yes it does, there is strong positive correlation between number of healthcare worker(HCW) cases and comorbidity cases

```python
#Create scatter plot to explore the relationship
plt.figure(figsize=(10,6))
sns.scatterplot(x='booster_hcw', y='booster_comorbid', data=data, color='red')
plt.title('Correlation between HCW and Comorbidity Cases')
plt.xlabel('Number of Booster HCW Cases')
plt.ylabel('Number of Booster Comorid Cases')
plt.show()
```

#### Q5: How has the number of cases changed over time?

Based on the graph, we can see that comorbidity cases is peak on July 2021. We can see the trend where the value of cases until the end of 2022, then the value increase drastically on January 2022. After that, the cases decreases slowly

```python
#convert the 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])

#Group by date and sum the total number of cases
cases_over_time= data.groupby('date')['total_comorbid'].sum()

#plot line chart
cases_over_time.plot(kind='line', color='brown')
plt.title('Change of Comorbidity Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Number of Cases')
plt.show()
```
