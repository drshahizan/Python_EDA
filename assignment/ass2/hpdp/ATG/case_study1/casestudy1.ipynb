## Exploratory Analysis & Visualization
Exploratory analysis and visualization involve examining data through charts and visuals to discover insights, patterns, and trends for decision-making.

> Import the visualization libraries (Plotly Express, Seaborn, and Matplotlib).

```python
import plotly.express as px
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

> Use `dataset.describe()` to determine the statistical summary of the dataset's numeric columns. For example, mean (average), standard deviation (measure of data spread), minimum, 25th percentile, median (50th percentile), 75th percentile, and maximum values for each numeric column in the dataset.

### 1. Histogram for Population Distribution:
>

```python
dataset.groupby("state")["population"].sum().sort_values()
```

```python
district_population = dataset.groupby("state")["population"].sum().sort_values()

plt.figure(figsize=(15, 4))
district_population.plot(kind='bar', color='skyblue')
plt.title('Total Population by District')
plt.xlabel('State')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### 2. Bar Chart for Ethnicity Distribution

```python
plt.figure(figsize=(12, 4))
sns.countplot(x='ethnicity', data=dataset, palette='viridis')
plt.title('Ethnicity Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### 3. Stacked Bar Chart for Ethnicity and Gender Distribution

```python
ethnicity_gender = dataset.groupby(['ethnicity', 'sex']).size().unstack()
ethnicity_gender.plot(kind='bar', stacked=True, figsize=(16, 4), colormap='coolwarm')
plt.title('Ethnicity and Gender Distribution')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

### 4. Line Plot for Population Trends Over Time

```python
dataset['date'] = pd.to_datetime(dataset['date'])
plt.figure(figsize=(12, 5))
sns.lineplot(x='date', y='population', data=dataset, estimator='mean', ci=None)
plt.title('Population Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Mean Population')
plt.show()
```

### 5. Correlation Heatmap for Numeric Columns

```python
numeric_columns = dataset.select_dtypes(include='number')
correlation_matrix = numeric_columns.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Numeric Columns')
plt.show()
```

### 6. Pair Plot for Multivariate Relationships (Age, Population, Sex, Ethnicity)

```python
sns.set_theme(style="ticks")
sns.pairplot(dataset[['age', 'population', 'sex', 'ethnicity']], hue='sex', markers=["o", "s"], palette='husl')
plt.suptitle('Pair Plot for Age, Population, Sex, and Ethnicity')

plt.show()
```

### 7. Pie Chart for Ethnicity Distribution in a Specific District

```python
specific_district = 'Johor Bahru'
district_data = dataset[dataset['district'] == 'Johor Bahru']

plt.figure(figsize=(10, 10))
plt.pie(district_data['ethnicity'].value_counts(), labels=district_data['ethnicity'].unique(), autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title(f'Ethnicity Distribution in {specific_district} District')
plt.show()
```

## Asking & Answering Questions About The Data

#### Q1. What is the distribution of population across different states in the dataset?

```python
state_population_distribution = dataset.groupby('state')['population'].sum().sort_values(ascending=False)
state_population_distribution
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x=state_population_distribution.index, y=state_population_distribution.values, palette='viridis')
plt.title('Population Distribution Across States')
plt.xlabel('State')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')
plt.show()
```

> The Bar Plot above has clearly shown the population distribution across each states in Malaysia. <br> For example, Selangor have the highest population and W.P. Labuan have the least population.

#### Q2. How does the population vary among different ethnicities in Johor state?

```python
specific_district_data = dataset[dataset['state'] == 'Johor']
ethnicity_distribution_in_district = specific_district_data.groupby('ethnicity')['population'].sum().sort_values(ascending=False)
ethnicity_distribution_in_district
```

```python
plt.figure(figsize=(6, 6))
plt.pie(ethnicity_distribution_in_district, labels=ethnicity_distribution_in_district.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Ethnicity Distribution in Johor State')
plt.show()
```

> The pie chart above shows that **bumi_malay** have the highest population in Johor whereas **other_citizen** have the least population in Johor.

#### Q3. What is the gender distribution within each ethnicity?

```python
gender_distribution_within_ethnicity = dataset.groupby(['ethnicity', 'sex'])['population'].sum().unstack()
gender_distribution_within_ethnicity
```

```python
plt.figure(figsize=(14, 8))
gender_distribution_within_ethnicity.plot(kind='area', stacked=True, colormap='viridis')
plt.title('Gender Distribution Within Each Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Total Population')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.show()
```

> The stacked chart above examines the different gender distribution by each ethnicity in Malaysia

#### Q4. How has the population distribution among different states changed over time?

```python
population_distribution_across_states_over_time = dataset.groupby(['date', 'state'])['population'].sum().unstack()
population_distribution_across_states_over_time
```

```python
plt.figure(figsize=(16, 8))
population_distribution_across_states_over_time.plot(kind='line', colormap='tab20', marker='o')
plt.title('Population Distribution Across States Over Time')
plt.xlabel('Date')
plt.ylabel('Total Population')
plt.legend(title='State', bbox_to_anchor=(1, 1))
plt.show()
```

> The line graph above shows that the difference of population distribution across states over the year from 2020-2023.

#### Q8. How does the population of each district in Pahang state compare to the overall state populationon1/1/2021?

```python
state_of_interest = 'Pahang'
date_of_interest = '1/1/2021'

# Filter the dataset for the specified state and date
filtered_data = dataset[(dataset['state'] == state_of_interest) & (dataset['date'] == date_of_interest)]

# Check if the filtered dataset is not empty
if not filtered_data.empty:
    # Calculate the population for each district on the specified date
    district_population = filtered_data.groupby('district')['population'].sum().sort_values(ascending=False)

    # Calculate the overall state population on the specified date
    overall_state_population = filtered_data['population'].sum()

    # Plot a bar chart to compare district population to overall state population
    plt.figure(figsize=(12, 8))
    sns.barplot(x=district_population.index, y=district_population.values, palette='viridis')
    plt.axhline(y=overall_state_population, color='r', linestyle='--', label=f'Overall State Population ({overall_state_population:,})')
    plt.xlabel('District')
    plt.ylabel('Population')
    plt.title(f'Population Comparison of Districts in {state_of_interest} on {date_of_interest}')
    plt.legend()
    plt.xticks(rotation=45, ha='right')
    plt.show()
else:
    print(f"No data available for {state_of_interest} on {date_of_interest}.")
```

> Based on the output above, there is no available data for Pahang state at 1/1/2021 to be compared to the overall dataset.
