### Data Type Preprocessing


To make data more comprehensive, needed to remove the 'RM' and 'per month' in ' monthly_rent' column and convert it from object to float.

```python
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].str.replace('RM', '').str.replace('per month', '').str.strip()
print(KL_rental_price['monthly_rent'])
```

Remove "sq.ft." for the size column.

```python
KL_rental_price['size'] = KL_rental_price['size'].str.replace('sq.ft.', '').str.strip()
print(KL_rental_price['size'])
```

Check the data type of every columns.

```python
KL_rental_price.dtypes
```

```python
KL_rental_price.info()
```

Convert the data type to the correct categories
- ads_id = int64 convert to object
- monthly_rent = object convert to float64
- size = object convert to float64

```python
# Replace 'More than 10' with NaN in the 'rooms' column
KL_rental_price['rooms'] = KL_rental_price['rooms'].replace('More than 10', np.nan)

# Drop rows with missing values in the 'rooms' column
KL_rental_price = KL_rental_price.dropna(subset=['rooms'])

# Convert 'rooms' to float (to handle values with decimal points)
KL_rental_price.loc[:, 'rooms'] = KL_rental_price['rooms'].astype(float)

# Convert 'rooms' to integers
KL_rental_price.loc[:, 'rooms'] = KL_rental_price['rooms'].astype(int)

# Check the data types
print(KL_rental_price.dtypes)
```

```python
# Convert 'ads_id' to object
KL_rental_price['ads_id'] = KL_rental_price['ads_id'].astype('object')

# Clean and convert 'monthly_rent' to float (remove spaces)
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].astype(str)
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].str.replace(' ', '', regex=True).astype(float)

# Convert 'size' to int
KL_rental_price.loc[:,'size'] = KL_rental_price['size'].astype(int)

# Convert 'bathroom' to int
KL_rental_price.loc[:,'bathroom'] = KL_rental_price['bathroom'].astype(int)

# Check the data types
print(KL_rental_price.dtypes)
```

## Duplicate Values
Check for the duplicate value.

```python
# Check for duplicate columns
duplicate_rows = KL_rental_price[KL_rental_price.duplicated()]


if not duplicate_rows.empty:
    print("Duplicate rows:")
    print(duplicate_rows)
    print(f"Total duplicate rows: {len(duplicate_rows)}")
else:
    print("No duplicate rows found.")
```

Remove for the duplicate value.

```python
# Remove duplicate rows
KL_rental_price = KL_rental_price.drop_duplicates()
```

### Missing Values

Checking and count for missing value and null value.




```python
null_values = KL_rental_price.isnull().sum()
null_values
```

```python
# Fill 'facilities' with "Not Available"
KL_rental_price['facilities'].fillna("Not Available", inplace=True)

# Fill 'additional_facilities' with "Not Available"
KL_rental_price['additional_facilities'].fillna("Not Available", inplace=True)

# Fill 'parking' with zero
KL_rental_price['parking'].fillna(0, inplace=True)

# Remove rows with missing values in 'monthly_rent'
KL_rental_price.dropna(subset=['monthly_rent'], inplace=True)

# Remove rows with missing values in 'rooms'
KL_rental_price.dropna(subset=['rooms'], inplace=True)

# Remove rows with missing values in 'furnished'
KL_rental_price.dropna(subset=['furnished'], inplace=True)

# Remove rows with missing values in 'completion_year'
KL_rental_price.dropna(subset=['completion_year'], inplace=True)

# Remove rows with missing values in 'prop_name'
KL_rental_price.dropna(subset=['prop_name'], inplace=True)
```

```python
# Replace rows where 'size' is 99999999 with NaN (missing value)
KL_rental_price['size'] = KL_rental_price['size'].replace(99999999, float('nan'))

# Drop rows with NaN values in the 'size' column
KL_rental_price = KL_rental_price.dropna(subset=['size'])

# Check the DataFrame after filling missing values
KL_rental_price.head(5)
```

#### Q6: What is the monthly rental price range for each property type?
Below chart shows as Autoviz:



```python
# Group by 'property_type' and calculate the rental price statistics
price_range = KL_rental_price.groupby('property_type')['monthly_rent'].agg(['min', 'max', 'mean'])

# Rename the columns for clarity
price_range.columns = ['min_rent', 'max_rent', 'average_rent']

# Print the monthly rental price statistics for each property type
print(price_range)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a stacked bar chart for rent range
fig, ax = plt.subplots(figsize=(9, 3))
property_types = price_range.index
rent_range = price_range['max_rent'] - price_range['min_rent']

# Create the bars for 'Rent Range'
rent_range_bars = plt.bar(property_types, rent_range, label='Rent Range', alpha=0.7)

plt.xlabel('Property Type')
plt.ylabel('Monthly Rental Range')
plt.title('Monthly Rental Price Range by Property Type')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

###Q9: How many units have at least 3 rooms and 2 bathrooms, and what is their monthly rental price range?

```python
# Filter the dataset for units with at least 3 rooms and 2 bathrooms
filtered_units = KL_rental_price[(KL_rental_price['rooms'] >= 3) & (KL_rental_price['bathroom'] >= 2)]

# Calculate the monthly rental price range for the filtered units
min_price = filtered_units['monthly_rent'].min()
max_price = filtered_units['monthly_rent'].max()

# Count the number of units that meet the criteria
unit_count = len(filtered_units)

# Print the results
print(f"Number of units with at least 3 rooms and 2 bathrooms: {unit_count}")
print(f"Monthly rental price range for these units: RM {min_price} to RM {max_price}")
```
