### Data Type Preprocessing


To make data more comprehensive, needed to remove the 'RM' and 'per month' in ' monthly_rent' column and convert it from object to float.

```python
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].str.replace('RM', '').str.replace('per month', '').str.strip()
print(KL_rental_price['monthly_rent'])
```

Remove "sq.ft." for the size column.

```python
KL_rental_price['size'] = KL_rental_price['size'].str.replace('sq.ft.', '').str.strip()
print(KL_rental_price['size'])
```

Check the data type of every columns.

```python
KL_rental_price.dtypes
```

```python
KL_rental_price.info()
```

Convert the data type to the correct categories
- ads_id = int64 convert to object
- monthly_rent = object convert to float64
- size = object convert to float64

```python
# Convert 'ads_id' to object
KL_rental_price['ads_id'] = KL_rental_price['ads_id'].astype('object')

# Clean and convert 'monthly_rent' to float (remove spaces)
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].astype(str)  # Ensure the column is treated as a string
KL_rental_price['monthly_rent'] = KL_rental_price['monthly_rent'].str.replace(' ', '', regex=True).astype(float)

# Convert 'size' to int
KL_rental_price['size'] = KL_rental_price['size'].astype(int)

# Convert 'rooms' to int
KL_rental_price['rooms'] = KL_rental_price['size'].astype(int)

# Convert 'bathroom' to int
KL_rental_price['bathroom'] = KL_rental_price['size'].astype(int)

# Check the data types
print(KL_rental_price.dtypes)
```

### Missing Values

Checking and count for missing value and null value.




```python
null_values = KL_rental_price.isnull().sum()
null_values
```

```python
# Fill 'facilities' with "Not Available"
KL_rental_price['facilities'].fillna("Not Available", inplace=True)

# Fill 'additional_facilities' with "Not Available"
KL_rental_price['additional_facilities'].fillna("Not Available", inplace=True)

# Fill 'parking' with zero
KL_rental_price['parking'].fillna(0, inplace=True)

# Remove rows with missing values in 'monthly_rent'
KL_rental_price.dropna(subset=['monthly_rent'], inplace=True)

# Remove rows with missing values in 'room'
KL_rental_price.dropna(subset=['rooms'], inplace=True)

# Remove rows with missing values in 'furnished'
KL_rental_price.dropna(subset=['furnished'], inplace=True)

# Check the DataFrame after filling missing values
KL_rental_price.head(5)
```

## Asking and Answering Questions

To conduct a more extensive exploration of available rentals with 10 questions below.



#### Q1: How many type of property type and list out the property type.

```python
import pandas as pd

# Extract and list unique property types
unique_property_types = KL_rental_price['property_type'].unique()
num_property_types = len(unique_property_types)

print(f"Number of unique property types: {num_property_types}")
print("Unique property types:")
for prop_type in unique_property_types:
    print(prop_type)
```

#### Q2: How many type of furnishing status and list out the status.

```python
# Use the unique() function to get the unique values in the 'furnished' column
furnishing_status = KL_rental_price['furnished'].unique()

# Count the number of unique furnishing status
num_status = len(furnishing_status)

# Print the number and list of furnishing status
print(f"There are {num_status} types of furnishing status.")
print("Furnishing status:")
for status in furnishing_status:
    print(status)
```

#### Q3: How many rental units available by each property type?

```python
# Group by property type and count the number of units for each type
property_type_counts = KL_rental_price['property_type'].value_counts()

# Display the result
print(property_type_counts)
```

#### Q4: How many rental units available by each property location?

```python
location_counts = KL_rental_price['location'].value_counts()
print(location_counts)
```

#### Q5: Make 2 columns with property name and age of property as of now (year 2023).

```python
# Calculate the age of each property
current_year = 2023  # Change to the current year
KL_rental_price['age_of_property'] = current_year - KL_rental_price['completion_year']

# Select the 'prop_name' and 'age_of_property' columns
property_age = KL_rental_price[['prop_name', 'age_of_property']]

print(property_age)
```

#### Q6: What is the monthly rental price range for each property type?

```python
# Group by 'property_type' and calculate the rental price range
price_range = KL_rental_price.groupby('property_type')['monthly_rent'].agg(['min', 'max'])

# Rename the columns for clarity
price_range.columns = ['min_rent', 'max_rent']

# Print the monthly rental price range for each property type
print(price_range)
```

###Q7:What is the size range for each property type?

```python
# Group the data by 'property_type' and calculate the size range
property_type_size_range = KL_rental_price.groupby('property_type')['size'].agg(['min', 'max'])

# Rename the columns for clarity
property_type_size_range.columns = ['Min Size', 'Max Size']

# Display the size range for each property type
print(property_type_size_range)
```

###Q8: What is the monthly rental price range for each property location?

```python
# Group by 'property location' and calculate the min and max of 'monthly_rent' for each group
rental_price_range = KL_rental_price.groupby('location')['monthly_rent'].agg(['min', 'max'])

# Rename the columns for clarity
rental_price_range = rental_price_range.rename(columns={'min': 'Min Monthly Rent', 'max': 'Max Monthly Rent'})

# Display the rental price range for each property location
print(rental_price_range)
```

###Q9: How many units have at least 3 rooms and 2 bathrooms, and what is their monthly rental price range?

```python
# Filter the dataset for units with at least 3 rooms and 2 bathrooms
filtered_units = KL_rental_price[(KL_rental_price['rooms'] >= 3) & (KL_rental_price['bathroom'] >= 2)]

# Calculate the monthly rental price range for the filtered units
min_price = filtered_units['monthly_rent'].min()
max_price = filtered_units['monthly_rent'].max()

# Count the number of units that meet the criteria
unit_count = len(filtered_units)

# Print the results
print(f"Number of units with at least 3 rooms and 2 bathrooms: {unit_count}")
print(f"Monthly rental price range for these units: RM {min_price} to RM {max_price}")
```

###Q10: What is the relationship between size of unit and its monthly rental price?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))
plt.scatter(KL_rental_price['size'], KL_rental_price['monthly_rent'], alpha=0.5)
plt.title('Relationship between Unit Size and Monthly Rental Price')
plt.xlabel('Size (square feet)')
plt.ylabel('Monthly Rental Price (RM)')
plt.grid(True)
plt.show()
```
