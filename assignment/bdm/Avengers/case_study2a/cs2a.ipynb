## Initialising the libraraies needed

```python
import sweetviz as sv
import pandas as pd
import numpy as np
import quickda

from dataprep.datasets import load_dataset
from dataprep.eda import create_report
```

### Importing the Dataset

```python
resturants_data = pd.read_csv("https://raw.githubusercontent.com/BakungaBronson/Trip-Advisor-Dataset/main/TripAdvisor_data_cleaned.csv")
resturants_data
```

## Report generation
Sweetviz comes with a function that can perform Exploratory Data Analysis and output and HTML report for us. Run the commands below to generate and display the report for our dataset.

```python
my_report = sv.analyze(resturants_data)
my_report.show_html(open_browser=True) # Default arguments will generate to "SWEETVIZ_REPORT.html"
my_report.show_notebook(w='100%')
```

## Pros and Cons Analysis

### SweetViz

- **Pros**:
  - Easy to install and use.
  - Simple visualizations with summaries at the top of the report.
- **Cons**:
  - May be too summarized making it hard to draw conclusions for very large datasets.
